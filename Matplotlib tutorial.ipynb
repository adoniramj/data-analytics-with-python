Creating simple plots in matplotlib
==

## Required modules

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

## Simple plot. No frills.

```python
# create a range of even numbers
evens = range(2,102,2)
plt.plot(evens)
plt.show()
```

In the code above, since only one parameter is specified, Pyplot assumes the input is for the y axis and it will choose a scale for the x axis.

## Plotting with a legend

```python
evens = range(2,102,2)
evens = np.array(evens)
plt.plot(evens, evens ** 2, label = 'x^2')
plt.legend()
plt.show()
```

## Saving a figure

```python
x = np.arange(1,11)
y = x**3
plt.plot(x, y, label = 'x^3')
plt.legend()
plt.savefig('x-cube.png', dpi = 300) # this function must come before show()
plt.show()
```
